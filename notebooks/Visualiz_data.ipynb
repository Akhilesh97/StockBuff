```python
def remove_non_ascii(string):
    if type(string) == str:
        return ''.join(char for char in string if ord(char) < 128)
    else:
        return string
```

```python
news_df_li = []
for file in os.listdir("../data/datasets/news_data/"):
    news_df_li.append(pd.read_csv("../data/datasets/news_data/%s"%file))
```

```python
news_df = pd.concat(news_df_li)
```

```python
news_df["tweet"] = news_df["tweet"].apply(remove_non_ascii)
```
